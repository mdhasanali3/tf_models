```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train , _ ), (x_test, _)= mnist.load_data()
```

```python
print(x_train)
print(x_train.shape)
print(x_train.dtype)
```

```python
print(x_test)
print(x_test.shape)
print(x_test.dtype)
```

```python
x_train = x_train.astype('float32')/255.
x_test = x_test.astype('float32')/255.
print(x_train.dtype)
x_train = np.reshape(x_train,(len(x_train),28,28,1))
x_test= np.reshape(x_test, (len(x_test),28,28,1))
print(x_train.dtype)
```

```python
print(x_train)
print(x_train.shape)
print(x_train.dtype)
```

```python
print(x_test)
print(x_test.shape)
print(x_test.dtype)
```

```python
#adding artificial noise
#random noise from normal distribution with mean at 0 and std dev of 1      
noise_factor=.5
x_train_noisy= x_train + noise_factor * np.random.normal(loc= 0.0, scale =1.0, size= x_train.shape)
x_test_noisy = x_test+ noise_factor * np.random.normal(loc=0.0, scale=1.0, size= x_test.shape)
```

```python
#clip all values between 0 and 1
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

```python
# display image with noise
plt.figure(figsize=(20,2))
for i in range(1,10):
  ax=plt.subplot(1,10,i)
  plt.imshow(x_test_noisy[i].reshape(28,28), cmap='binary')

plt.show()

```

```python
model= Sequential()

#down sample
model.add(Conv2D(32, (3,3), activation='relu', padding ='same', input_shape=(28,28,1)))
model.add(MaxPooling2D((2,2),padding='same'))

model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D((2,2), padding='same'))

model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D((2,2),padding='same'))


# up sample
model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
model.add(UpSampling2D((2,2)))

model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
model.add(UpSampling2D((2,2)))

model.add(Conv2D(32, (3,3), activation='relu'))
model.add(UpSampling2D((2,2)))

model.add(Conv2D(1, (3,3), activation ='relu', padding ='same'))

model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train_noisy, x_train, epochs=10, batch_size= 256, shuffle=True , 
          validation_data=(x_test_noisy,x_test))
```

```python
model.evaluate(x_test_noisy, x_test)
```

```python
no_noise_img= model.predict(x_test_noisy)
```

```python
plt.figure(figsize=(40,4))
for i in range(10):
  #display original
  ax= plt.subplot(3, 20, i+1)
  plt.imshow(x_test_noisy[i].reshape(28,28), cmap='binary')

  #display reconstructed (after noise removed) image
  ax=plt.subplot(3, 20, 40 +i+1)
  plt.imshow(no_noise_img[i].reshape(28,28), cmap='binary') 

plt.show()

```
